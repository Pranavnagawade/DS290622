```python
class TicketInfo:
    def __init__(self, row, col):
        self.row = row
        self.col = col
        self.user_details = {}
    
    def options(self, option):
        if option == 1:
            self.show_seats()
            
        elif option == 2:
            self.buy_ticket()
            
        elif option == 3:
            self.statistics()
            
        elif option == 4:
            self.user_info()
        else:
            print('Enter a valid Option')
            
            
    def show_seats(self):
        for i in range(self.row + 1):
            for j in range(self.col + 1):
#                 row 0
                if i == 0:
#                   col 0
                    if j == 0:
                        print(' ', end = ' ')
#                   row 0 and col 1 -> n
                    else:
                        print(j, end = ' ')
                else:
#                   col 0 and row 1 -> n
                    if j == 0:
                        print(i, end = ' ')
#                   row and col != 0
                    else:
                        if self.seat_status(i, j):
                            print('B', end = ' ')
                        else:
                            print('S', end = ' ')
            print()
    
    
    def buy_ticket(self):
        row = int(input("For which row you are planning to book the ticket: "))
        col = int(input("For which column you are planning to book the ticket: "))
        seat_id = str(row) + str(col)
        total_seats = self.row * self.col
        half_seats = total_seats // 2
        if total_seats < 60 or (total_seats > 60 and row <= half_seats):
            price = 10
        else:
            price = 8
        if self.seat_status(row, col):
            print("That seat is already reserved")
        else:
            print('Price of the selected seat is', price, '$')
            choice = input('Enter Yes to continue to purchase the ticket')
            if choice.lower() == 'yes':
                name = input('Enter your Name')
                gender = input('Enter your Gender')
                age = input('Enter your Age')
                mobile_number = input('Enter your Mobile Number')
                self.user_details[seat_id] = [name, gender, age, mobile_number, price]
                print('Your ticket is booked your seating id is', seat_id)
#                 self.user_details[seat_id] = name + gender + age + mobile_number
            else:
                print('No problem visit back soon...........')
    
    
    def statistics(self):
        reserved_tickets = len(self.user_details)
        percentage = (reserved_tickets / (self.row * self.col)) * 100
        total = 0
#         {'22': [name, gend,age, mob, 10]}
        for price in self.user_details.values():
            total += price[4]
        seats = self.row * self.col
        income = 0
        if seats <= 60:
            income = seats * 10
        else:
            half_seats = (self.row // 2) * self.col
            income = half_seats * 10 + (seats - half_seats) * 8 
#             9 8 4 5
        s = 'Number of Purchased Tickets {} \nPercentage of Tickets booked {} \nCurrent Income {} \nTotal Income{}'.format(reserved_tickets, percentage, total, income)
        print(s)
        
    def user_info(self):
        pass
    def seat_status(self, row, col):
        s_id = str(row)+str(col)
        if s_id in self.user_details.keys():
            return True
        False
    
opt = """1. Show the seats
2. Buy a Ticket
3. Statistics
4. Show booked Tickets User Info
0. Exit\n"""



row = int(input('How many rows avaliable in theater'))
col = int(input('How many columns avaliable in theater'))
Admin = TicketInfo(row, col)
while True:
    user_option = int(input(opt))
    if user_option == 0:
        break
    else:
        Admin.options(user_option)
```
